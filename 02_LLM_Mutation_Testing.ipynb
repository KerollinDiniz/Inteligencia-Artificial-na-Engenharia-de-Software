```python
%%writefile pom.xml
<project xmlns="http://maven.apache.org/POM/4.0.0"
         xmlns:xsi="http://www.w3.org/2001/XMLSchema-instance"
         xsi:schemaLocation="http://maven.apache.org/POM/4.0.0
         http://maven.apache.org/xsd/maven-4.0.0.xsd">

    <modelVersion>4.0.0</modelVersion>
    <groupId>com.exemplo</groupId>
    <artifactId>projeto-llm</artifactId>
    <version>1.0-SNAPSHOT</version>

    <properties>
        <maven.compiler.source>17</maven.compiler.source>
        <maven.compiler.target>17</maven.compiler.target>
    </properties>

    <dependencies>
        <dependency>
            <groupId>org.junit.jupiter</groupId>
            <artifactId>junit-jupiter</artifactId>
            <version>5.9.3</version>
            <scope>test</scope>
        </dependency>
    </dependencies>

    <build>
        <plugins>

            <plugin>
                <groupId>org.apache.maven.plugins</groupId>
                <artifactId>maven-surefire-plugin</artifactId>
                <version>3.0.0</version>
            </plugin>

            <plugin>
                <groupId>org.pitest</groupId>
                <artifactId>pitest-maven</artifactId>
                <version>1.15.0</version>
                <configuration>
                    <targetClasses>
                        <param>com.exemplo.*</param>
                    </targetClasses>
                    <targetTests>
                        <param>com.exemplo.*</param>
                    </targetTests>
                </configuration>
                <dependencies>
                    <dependency>
                        <groupId>org.pitest</groupId>
                        <artifactId>pitest-junit5-plugin</artifactId>
                        <version>1.2.0</version>
                    </dependency>
                </dependencies>
            </plugin>

        </plugins>
    </build>

</project>
```

```python
%%writefile src/main/java/com/exemplo/CalculadoraDesconto.java
package com.exemplo;

public class CalculadoraDesconto {

    public double calcularDesconto(double valor, boolean vip) {

        if (valor <= 0) {
            throw new IllegalArgumentException("Valor invÃ¡lido");
        }

        if (vip) {
            if (valor > 1000) {
                return valor * 0.8;
            }
            return valor * 0.95;
        } else {
            if (valor > 1000) {
                return valor * 0.9;
            }
            return valor;
        }
    }
}
```

```python
%%writefile src/test/java/com/exemplo/CalculadoraDescontoLLMTest.java
package com.exemplo;

import org.junit.jupiter.api.Test;
import static org.junit.jupiter.api.Assertions.*;

public class CalculadoraDescontoLLMTest {

    @Test
    void deveAplicar20PorCentoParaVipCompraAlta() {
        CalculadoraDesconto calc = new CalculadoraDesconto();
        assertEquals(1600.0, calc.calcularDesconto(2000, true));
    }

    @Test
    void deveAplicar10PorCentoParaNaoVipCompraAlta() {
        CalculadoraDesconto calc = new CalculadoraDesconto();
        assertEquals(1800.0, calc.calcularDesconto(2000, false));
    }

    @Test
    void deveAplicar5PorCentoParaVipCompraBaixa() {
        CalculadoraDesconto calc = new CalculadoraDesconto();
        assertEquals(760.0, calc.calcularDesconto(800, true));
    }

    @Test
    void naoDeveAplicarDescontoParaNaoVipCompraBaixa() {
        CalculadoraDesconto calc = new CalculadoraDesconto();
        assertEquals(800.0, calc.calcularDesconto(800, false));
    }
}
```
